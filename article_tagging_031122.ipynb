```python
import pandas as pd
import numpy as np
```

```python
#MARINA FRIDMAN'S CODE
# load data (prepared by load_climate_clusts)
df = pd.read_csv('../data/klima_articles_5.csv', index_col=0)

# convert published date to datetime
date_format = '%Y-%m-%dT%H:%M:%S'
df['pub_date_dt'] = pd.to_datetime(df.pub_date, format=date_format)

# find NTB articles
df['is_ntb'] = df.authors.str.contains('NTB')
df['is_ntb'] = df.is_ntb.fillna(False)

# fix indices
df['new_idx'] = np.where(df.id.isna(), df.index, df.id)
df.drop('id',axis=1, inplace=True)
df.set_index('new_idx', inplace=True)
df.index.rename('id', inplace=True)

# drop duplicates
df = df[(~df.index.duplicated())]

# fill in nans
df.description.fillna('',inplace=True)
df.content.fillna('',inplace=True)
```

# Balance random data
We want to have around 20 articles for each topic of interest.

The topics are :
+ Økonomi, næringsliv og innovasjon 
+ Politikk, regjering og storting 
+ Kosthold, helse og livsstil 
+ Klima og norsk politikk 
+ Klima og internasjonal politikk 
+ Klima og transport 
+ Klima og fornybar energi 
+ Klima og olje- og gassindustrien 
+ Klima, mat og landbruk + diet/agriculture 
+ natural disasters 
+ Fisheries 
+ Forest health (needs better name, swamps should also go there) 

The articles should capture words that can be but not limited to the words in the category label. They should be as heterogenous as possible. Try for 5-10 keywords, and about 5 articles per keyword.

```python
df_supp = pd.concat((df[df.description.str.lower().str.contains('gass')].sample(20),
            df[df.description.str.lower().str.contains('oljeindustrien')].sample(20),
            df[df.description.str.lower().str.contains('equinor')].sample(20),
            df[df.description.str.lower().str.contains('equinor')].sample(20)))

# save articles
df_supp.to_csv('../data/non_random_climate_articles.csv')
```
